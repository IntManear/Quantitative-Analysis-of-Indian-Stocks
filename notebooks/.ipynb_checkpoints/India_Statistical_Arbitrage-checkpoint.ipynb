```python
list_of_etfs = ['ACC','ADANIPORTS','AMBUJACEM','ASIANPAINT','AXISBANK','BANKBARODA','BHEL','BPCL', 'BOSCHLTD','CIPLA', 'COALINDIA', 'DRREDDY',                     'GAIL','GRASIM','HCLTECH','HDFCBANK','HEROMOTOCO','HINDALCO','HINDUNILVR', 'HDFC','ITC', 'ICICIBANK', 'IDEA', 'INDUSINDBK','INFY',                  'KOTAKBANK','LT','LUPIN', 'MARUTI','NTPC', 'ONGC', 'POWERGRID','PNB','RELIANCE','SBIN','SUNPHARMA','TCS','TATAMOTORS', 'TATAPOWER',                 'TATASTEEL', 'TECHM', 'ULTRACEMCO', 'VEDL','WIPRO','YESBANK','ZEEL']
left = ['BHARTIARTL']
for etf in list_of_etfs:
    try:
        exec('{} = pd.DataFrame()'.format(etf))
        exec('{} = pdr.get_data_yahoo("{}"+".NS",start)'.format(etf,etf)) 
    except:
        print("Data not available for", etf)
```

```python
corr_data = dict()
for i in range(len(list_of_etfs)):
  for j in range(i+1,len(list_of_etfs)):
    if(i==j):
      pass
    else:
      stocks = [list_of_etfs[i],list_of_etfs[j]]
      print(stocks)
      cointegrated_etfs_close_price = pd.DataFrame()
      # cointegrated_etfs_close_price = cointegrated_etfs_close_price.reset_index()
      for stock in stocks: 
        exec("cointegrated_etfs_close_price[stock] = {}['Close']".format(stock))
      # cointegrated_etfs_close_price['ACC'] = ACC['Close']
      # cointegrated_etfs_close_price['TATAPOWER'] = TATAPOWER['Close']
      print("The correlation between",stocks[0],"and",stocks[1],cointegrated_etfs_close_price.loc[:,[stocks[0],stocks[1]]].corr()[stocks[0]][stocks[1]]*100,"%")
      corr = cointegrated_etfs_close_price.loc[:,[stocks[0],stocks[1]]].corr()[stocks[0]][stocks[1]]*100
      corr_data.setdefault(stocks[0],{})[stocks[1]] = corr
```

```python
corr_data['ACC']['TATAPOWER']
```

```python
keys = corr_data.keys
```

```python
best = dict()
for i in range(len(list_of_etfs)):
  for j in range(len(list_of_etfs)):
      if(i>=j):
          pass
      else:
          stocks = [list_of_etfs[i],list_of_etfs[j]]
          value = corr_data[stocks[0]][stocks[1]]
          if (value>=95.0):
            best.setdefault(stocks[0],{})[stocks[1]] = value
          else:
            pass
```
