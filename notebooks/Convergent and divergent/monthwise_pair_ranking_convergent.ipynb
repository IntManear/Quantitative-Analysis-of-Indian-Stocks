```python
import datetime

def last_day_of_month(any_day):
    next_month = any_day.replace(day=28) + datetime.timedelta(days=4)  # this will never fail
    return next_month - datetime.timedelta(days=next_month.day)
```

```python
start_list = pd.date_range("2018-01-01","2019-12-31", 
              freq='MS').strftime("%Y-%m-01").tolist()
end_list = []
for month in range(1, 13):
  #index = (2*month) - 1
  date = str(last_day_of_month(datetime.date(2018, month, 1)))
  # date_list.(index, date)
  end_list.append(date)

for month in range(1, 13):
  #index = ((2*month) - 1) + 24
  date = str(last_day_of_month(datetime.date(2019, month, 1)))
  # date_list.insert(index, date)
  end_list.append(date)
```
