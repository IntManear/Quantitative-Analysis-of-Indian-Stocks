```python
import pandas as pd
import datetime
from tqdm import tqdm
```

```python
'''Getting names of all csv files(stocks) in the folder'''

import os
path = '/home/dhruv/Downloads/f&o iday data'
csv_files_initial = [f for f in os.listdir(path) if f.endswith('.CSV')]
```

```python
stocks_initial = [sub[ : -4] for sub in csv_files_initial]
```

```python
for i in range(len(stocks_initial)):
    print(csv_files_initial[i],'-->',stocks_initial[i])
```

```python
stocks = []   
csv_files = []
for i in range(len(csv_files_initial)):
    try:
        stock = stocks_initial[i]
        csv = csv_files_initial[i]
        #print(stock,csv)
        exec('{} = pd.read_csv("/home/dhruv/Downloads/f&o iday data/{}")'.format(stock,csv))
        stocks.append(stock)
        csv_files.append(csv)
    except:
        print(stock)
```

```python
for stock in stocks:
    exec('{}["Dates"],{}["Time"] = {}["Date"].str.split(",",1).str'.format(stock,stock,stock))
    exec('{}["Dates"] = pd.to_datetime({}["Dates"])'.format(stock,stock))
#ACC['Dates'], ACC['Time'] = ACC['Date'].str.split(',', 1).str
```

```python
list_of_names = []
for stock in stocks:
    month1 = str(stock) +'1'
    month2 = str(stock) +'2'
    month3 = str(stock) +'3'
    month4 = str(stock) +'4'
    month5 = str(stock) +'5'
    month6 = str(stock) +'6'
    temp_list = [month1,month2,month3,month4,month5,month6]
    exec('{},{},{},{},{},{} = [g for n, g in {}.groupby(pd.Grouper(key="Dates",freq="M"))]'.format(month1,month2,month3,month4,month5,month6,stock))
    for i in temp_list:
        exec('{}.reset_index(drop = True,inplace = True)'.format(str(i)))
    list_of_names.append(temp_list)
```

```python
'''Normalised close column created for every dataframe'''

for i in range(len(list_of_names)):
  for j in range(len(list_of_names[0])):
    name = list_of_names[i][j]
    print(name)
    exec('{}["Norm Close"] = {}["Close"] / {}["Close"][0]'.format(name,name,name))
```

```python
'''Difference between corresponding stocks calculated and stored in columns created in each dataframe'''

for m in tqdm(range(len(stocks))):
  for n in range(len(list_of_names[0])):
    i = list_of_names[m][n]

    for o in range(m+1,len(stocks)): 
        j = list_of_names[o][n]
        if(i==j):
          pass
        else:
          col_name = "diff_"+str(j)
          col2_name = "nonabs_diff_"+str(j)
          exec('{}["{}"] = {}["Norm Close"] - {}["Norm Close"]'.format(i,col_name,i,j))
          exec('{}["{}"] = {}["{}"].abs()'.format(i,col_name,i,col_name))
          exec('{}["{}"] = {}["{}"]'.format(i,col2_name,i,col_name))
```
